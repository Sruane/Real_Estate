# Merge dataframes

```python
#Merge df 1 with df 2 on indicator_id
data = data.merge(data_2,left_on='indicator_id',right_on='indicator_id' )
data.head()
```

```python
#Merge new df with df 3 on region_id
data = data.merge(data_3,left_on='region_id',right_on='region_id')
data.head()
```
